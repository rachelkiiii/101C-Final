```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("cleandata.csv")

# Count the number of positive and negative sentiments
sentiment_counts = df['sentiment'].value_counts()

# Create a pie chart for sentiment distribution
plt.figure(figsize=(8, 6))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Sentiment Distribution')
plt.show()

# Create a bar plot for sentiment counts
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['purple', 'orange'])
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Number of Positive and Negative Sentiments')
plt.xticks(rotation=0)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for emotions in positive, negative, and neutral sentiments
emotions = ['Joy', 'Anger', 'Sadness', 'Fear', 'Surprise']
positive_emotions = [0.8, 0.2, 0.1, 0.05, 0.3]
negative_emotions = [0.1, 0.6, 0.7, 0.3, 0.2]
neutral_emotions = [0.4, 0.3, 0.2, 0.1, 0.5]

# Number of emotions
num_emotions = len(emotions)

# Create radar chart
angles = np.linspace(0, 2 * np.pi, num_emotions, endpoint=False).tolist()

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, positive_emotions, color='green', alpha=0.25)
ax.fill(angles, negative_emotions, color='red', alpha=0.25)
ax.fill(angles, neutral_emotions, color='blue', alpha=0.25)

# Add labels for each spoke (emotion)
ax.set_xticks(angles)
ax.set_xticklabels(emotions)

plt.show() 
```
